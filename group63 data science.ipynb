```python
df = df_artisans.copy()
# Normalize rating (assuming 3–5 range)
df["rating_score"] = (df["Rating"] - 3) / (5 - 3)
# Normalize experience (1–25 years assumed)
df["experience_score"] = df["Years_Experience"] / 25
# Convert complaint rate into positive score
df["complaint_score"] = 1 - df["Complaint_Rate"]
# Availability mapping
availability_map = {
   "Available": 1,
   "Busy": 0.5,
   "Offline": 0
}
df["availability_score"] = df["Availability"].map(availability_map)
```

```python
from geopy.distance import geodesic
user_location = (6.5200, 3.3700)
def compute_distance(row):
   return geodesic(
       user_location,
       (row["Latitude"], row["Longitude"])
   ).km
df["distance_km"] = df.apply(compute_distance, axis=1)
# Normalize distance (closer = better)
max_distance = df["distance_km"].max()
df["distance_score"] = 1 - (df["distance_km"] / max_distance)
```

```python
requested_skill = "Electrician"
df_filtered = df[df["Skill_Category"] == requested_skill].copy()
```

```python
df_filtered["final_score"] = (
   0.30 * df_filtered["rating_score"] +
   0.20 * df_filtered["distance_score"] +
   0.20 * df_filtered["experience_score"] +
   0.15 * df_filtered["availability_score"] +
   0.15 * df_filtered["complaint_score"]
)
```

```python
import numpy as np
# Create booking probability using your ranking features
df_filtered["booking_probability"] = (
   0.4 * df_filtered["rating_score"] +
   0.3 * df_filtered["distance_score"] +
   0.3 * df_filtered["complaint_score"]
)
# Convert to 0/1 booking outcome
df_filtered["booked_successfully"] = np.random.binomial(
   1,
   df_filtered["booking_probability"]
)
```

```python
df_filtered = df_filtered.reset_index(drop=True)
```

```python
from sklearn.linear_model import LogisticRegression

from sklearn.model_selection import train_test_split

features = df_filtered[[

    "rating_score",

    "distance_score",

    "experience_score",

    "availability_score",

    "complaint_score"

]]

target = df_filtered["booked_successfully"]

X_train, X_test, y_train, y_test = train_test_split(

    features, target, test_size=0.2

)

model = LogisticRegression()

model.fit(X_train, y_train)

```
